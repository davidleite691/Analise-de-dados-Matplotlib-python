## Gráfico de barras empilhadas para mostrar a distribuição de diferentes tipos de vendas ao longo do tempo

Considere que você tenha uma empresa e queira mostrar a distribuição de diferentes tipos de vendas ao longo do tempo. Crie um gráfico de barras empilhadas:

```python
import matplotlib.pyplot as plt
from cycler import cycler

anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

plt.title("Vendas da loja online e da loja física")

# rc = run configuration (configuração quando executa)
# Todos os gráfios gerados serão em tom pastel
plt.rc('axes', prop_cycle=(cycler('color', plt.get_cmap('Pastel2').colors)))

plt.bar(anos, vendas_online, label = "Loja online")
plt.bar(anos, vendas_loja, bottom=vendas_online, label = "Loja física")

# forçar os anos a serem exibidos como inteiros
plt.xticks(ticks=anos)
plt.ylabel("Vendas (milhoes de R$)")
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
from cycler import cycler

anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

# rc = runtime configuration (configuração quando executa)
cores = plt.get_cmap('Pastel2').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)

barra_online = plt.bar(anos, vendas_online, label='Online')
barra_loja = plt.bar(anos, vendas_loja, bottom=vendas_online, label='Loja')
plt.bar_label(barra_online, labels=vendas_online, fontsize=10, fontweight='bold', label_type='center')
plt.bar_label(barra_loja, labels=vendas_loja, fontsize=10, fontweight='bold', label_type='center')

# forçar os anos a serem exibidos como inteiros
plt.xticks(ticks=anos)
plt.yticks([])
plt.tick_params(axis='x', length=0)
plt.box(False)

plt.ylabel('Vendas (em milhões)')
plt.title('Vendas Online vs Loja')
plt.legend()
plt.show()
```

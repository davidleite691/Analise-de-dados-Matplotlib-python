## Crie um gráfico de barras para comparar as temperaturas médias de diferentes cidades

Vamos supor que temos dados de temperatura média de três cidades: São Paulo, Rio de Janeiro e Salvador. Crie um gráfico de barras para comparar essas temperaturas:


```python
import matplotlib.pyplot as plt

cidades = ['São Paulo', 'Rio de Janeiro', 'Salvador']
temperaturas = [20, 25, 27]

barras = plt.bar(cidades,temperaturas,color = "green")
plt.title("Coparação de temperaturas médias de diferentes cidades")
plt.ylabel("Temperaturas em C°")
plt.xlabel("Cidades")
plt.bar_label(barras, labels = temperaturas)#colocas rótulo nas barras
plt.show()
```

## Gráfico de barras horizontais para as quantidades de produtos vendidos em uma loja

```python
import matplotlib.pyplot as plt
from cycler import cycler

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]
```

```python
dados_ordenados = sorted(zip(quantidades, produtos))  # ordenamento por quantidade

# Listas ordenadas dos prdutos
quantidades_ordenadas = [x[0] for x in dados_ordenados]
produtos_ordenados = [x[1] for x in dados_ordenados]
```

```python
cores = plt.get_cmap('Pastel2').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)

barras = plt.barh(produtos_ordenados, quantidades_ordenadas)

# Destacando o maior valor de vendas
plt.bar_label(barras, labels=[max(quantidades_ordenadas) if x == max(quantidades_ordenadas) else '' for x in quantidades_ordenadas],
              padding=3, fontsize=10, fontweight='bold')

# retirando as bordas do gráfico
plt.box(False)

# retirando o eixo x
plt.xticks([])
plt.tick_params(axis='y', length=0)

# destacando a barra do produto com maior quantidade vendida
for i, barra in enumerate(barras):
    if i != quantidades_ordenadas.index(max(quantidades_ordenadas)):
        barras[i].set_color(cores[1])
        barras[i].set_alpha(0.6)


plt.title('Quantidade de Produtos Vendidos')

plt.show()
```

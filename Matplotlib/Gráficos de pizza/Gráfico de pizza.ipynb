## Crie um gráfico de pizza para mostrar a distribuição de categorias de produtos em uma loja

Suponha que você tenha uma loja e queira mostrar a distribuição de categorias de produtos em sua loja. Você pode usar um gráfico de pizza para mostrar a distribuição de categorias de produtos em sua loja:


```python
import matplotlib.pyplot as plt

categorias = ['Eletrônicos', 'Roupas', 'Alimentos', 'Livros']
quantidades = [45, 30, 15, 10]  # valores em porcentagem
destaques = [0.1, 0, 0, 0]
```

```python
plt.title("Gráfico de distribuição de categorias de produtos")

cores = plt.get_cmap('Pastel2').colors

plt.pie(quantidades, autopct="%.1f%%", explode = destaques, colors=cores)
plt.legend(categorias,loc='center left', bbox_to_anchor=(1, 0.8))
plt.show()
```
